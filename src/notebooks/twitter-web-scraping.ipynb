```python
def collect_data(list_in_list, keyword_to_be_added=''):
    """Function to collect data from twitter which contains the keywords that are present in the list
    list_in_list: A String list which contains the keywords for which tweets must be collected
    returns a dataframe containing tweets and labels
    key_word_to_be_added - Keyword to be added to every topics in the list"""
    
    # Variables to store tweets and labels
    tweets, labels = [], []
    
    for inner_list in list_in_list:
        
        # To convert the string into lists
        split_list = inner_list.split(', ')
        for sub_label in split_list:
            for i,tweet in enumerate(sntwitter.TwitterSearchScraper(sub_label.strip() + keyword_to_be_added).get_items()):
                if i>5000:
                    break
                tweets.append(tweet.content)
                labels.append(split_list[0])
    df = pd.DataFrame({'tweets':tweets, 'labels':labels})
    return df
```

```python
%%time
complaints_df = collect_data(label_list, ' complaints India')
```
